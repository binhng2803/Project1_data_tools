```python
import os
```

```python
data_dir = os.listdir('C:\Workspace\Aimesoft\Building\Project1_data_tools\data\\fake_data')
image_files = [file for file in data_dir if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.jpeg')]
label_files = [file for file in data_dir if file.endswith('.json')]
len(image_files), len(label_files)
```

```python
dict_file = {file:[] for file in label_files}
for file in label_files:
    for img in image_files:
        if img.startswith(file[:-4]):
            dict_file[file].append(img)
wrong = [file for file in dict_file.keys() if len(dict_file[file])!=1]
wrong
```

```python
label_files = list(dict_file.keys())
image_files = [file[0] for file in dict_file.values()]
len(label_files), len(image_files)
```

```python
data_dir = 'C:\Workspace\Aimesoft\Building\Project1_data_tools\data\\fake_data'
img_dir = 'C:\Workspace\Aimesoft\Building\Project1_data_tools\data\\training_data\image'
label_dir = 'C:\Workspace\Aimesoft\Building\Project1_data_tools\data\\training_data\label'
import shutil
from tqdm import tqdm
for file in tqdm(image_files):
    shutil.copy(os.path.join(data_dir, file), img_dir)
for file in tqdm(label_files):
    shutil.copy(os.path.join(data_dir, file), label_dir)
```
